```python
from pyspark.sql import SparkSession

# Initialize SparkSession
spark = SparkSession.builder \
    .appName("SparkSessionInitialization") \
    .getOrCreate()

print("SparkSession initialized successfully!")
```

### Finding Maximum Salary in Employee Data

First, let's create a sample Spark DataFrame with employee data including names and salaries.

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

# Sample employee data
data = [("Alice", 50000),
        ("Bob", 60000),
        ("Charlie", 75000),
        ("David", 55000),
        ("Eve", 80000)]

# Define schema for the DataFrame
schema = StructType([
    StructField("Name", StringType(), True),
    StructField("Salary", IntegerType(), True)
])

# Create Spark DataFrame
emp_df = spark.createDataFrame(data, schema)

print("Employee DataFrame created:")
emp_df.show()
```

Now, let's find the maximum salary from this DataFrame.

```python
from pyspark.sql.functions import max

# Find the maximum salary
max_salary_df = emp_df.agg(max("Salary").alias("MaxSalary"))

print("Maximum Salary:")
max_salary_df.show()
```

The URL above points to the Spark UI. If it's a `localhost` address (e.g., `http://your-ip:4040`), you typically cannot access it directly from your browser in a cloud environment like Colab without setting up a tunnel.

Common approaches to access the Spark UI from Colab include:

*   **Port Forwarding/Tunneling:** Using tools like `ngrok` to expose the `localhost` port to a public URL. This usually requires installing `ngrok`, getting an auth token, and running a command to tunnel the port (e.g., `!ngrok http 4040`).
*   **Colab's Internal Proxies (if available):** Sometimes Colab provides an internal proxy mechanism, but it's not a standard feature for Spark UI.

After installing `ngrok`, you need to authenticate it with your `ngrok` auth token. If you don't have an `ngrok` account, you'll need to create one at [ngrok.com](https://ngrok.com) to get your auth token.

Once you have your auth token, replace `YOUR_NGROK_AUTH_TOKEN` with your actual token in the cell below and run it.

```python
from pyngrok import ngrok

# Authenticate ngrok (replace YOUR_NGROK_AUTH_TOKEN with your actual token)
# You can find your auth token on your ngrok dashboard: https://dashboard.ngrok.com/get-started/your-authtoken
ngrok.set_auth_token("<ngrok_auth_token>")
```

Now that `ngrok` is authenticated, we can start the tunnel for the Spark UI, which is running on port `4040` (as identified by `spark_ui_url`).

Run the cell below, and `ngrok` will provide a public URL you can use to access the Spark UI in your browser.

```python
from pyngrok import ngrok

# Start ngrok tunnel for Spark UI (port 4040)
# This will print a public URL that you can open in your browser
# Make sure you have set your ngrok auth token first.

try:
    # Disconnect any existing tunnels first to avoid conflicts
    ngrok.kill()

    # Connect to the Spark UI port (4040)
    tunnel = ngrok.connect(4040)
    print(f"Ngrok tunnel URL: {tunnel.public_url}")
    print("You can now open this URL in your browser to access the Spark UI.")
except Exception as e:
    print(f"Error starting ngrok tunnel: {e}")
    print("Please ensure your ngrok auth token is set correctly and try again.")
```
